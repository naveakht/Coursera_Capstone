```python
import pandas as pd
import numpy as np


web=pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')
pd.options.mode.chained_assignment = None  # default='warn'
```

# Clean the data 

```python
web_df = web[0]

#Removing NotAssigned Borough
web_df2= web_df[web_df['Borough'] != 'Not assigned']

web_df2.head()
```

```python
#Override Not Assigned Neighbourhood to Bourough
web_df2.loc[web_df2['Neighbourhood'] == 'Not assigned','Neighbourhood']=web_df2['Borough']
```

# Joining the Neighbourhood in seperate columns and removing the duplicate.

```python
web_df2['Neighbourhood_Con'] = web_df2.groupby(by=(['Postcode','Borough']))['Neighbourhood'].transform(lambda x: ','.join(x))

web_df2.drop(columns='Neighbourhood',inplace=True)
web_df2.drop_duplicates(inplace=True)
web_df2
```
