```python
import pandas as pd
import numpy as np


web=pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')
pd.options.mode.chained_assignment = None  # default='warn'
```

# Clean the data 

```python
web_df = web[0]

#Removing NotAssigned Borough
web_df2= web_df[web_df['Borough'] != 'Not assigned']

web_df2.head()
```

```python
#Override Not Assigned Neighbourhood to Bourough
web_df2.loc[web_df2['Neighbourhood'] == 'Not assigned','Neighbourhood']=web_df2['Borough']
```

# Joining the Neighbourhood in seperate columns and removing the duplicate.

```python
web_df2['Neighbourhood_Con'] = web_df2.groupby(by=(['Postcode','Borough']))['Neighbourhood'].transform(lambda x: ','.join(x))

web_df2.drop(columns='Neighbourhood',inplace=True)
web_df2.drop_duplicates(inplace=True)

web_df2.set_index('Postcode',inplace=True)
```

```python
web_df2.head()
```

```python
#printing the shape

web_df2.shape
```

# Finding Longitude and Latitude

```python
geo_data = pd.read_csv('https://cocl.us/Geospatial_data')
geo_data.head()
geo_data.set_index('Postal Code',inplace = True)
geo_data.head()
```

```python
for lPCCode in (web_df2.index):
    try:
        web_df2.loc[lPCCode,'Latitude'] = geo_data.loc[lPCCode,'Latitude']
        web_df2.loc[lPCCode,'Longitude'] = geo_data.loc[lPCCode,'Longitude']
    except:
        print('No data for ',lPCCode)

web_df2.head()
```

```python
web_df2.reset_index(inplace=True)

toronto_df=web_df2.loc[web_df2['Borough'].str.contains('Toronto')]
toronto_df
```

## Map toronto with Neighbour

```python
# create map of Toronto with Neighbour using latitude and longitude values

import folium
latitude=43.651070
longitude=-79.347015
toronto_map = folium.Map(location=[latitude, longitude], zoom_start=11)

# add markers to map
for lat, lng, label in zip(toronto_df['Latitude'], toronto_df['Longitude'], toronto_df['Neighbourhood_Con']):
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(toronto_map)  
    
toronto_map
```
